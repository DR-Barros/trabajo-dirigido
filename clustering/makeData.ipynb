```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Generamos un conjunto de datos de ejemplo

# Número de datos
n = 1000
dimensiones = 2

# Generamos n puntos distribuidos uniformemente en 2 grupos
# Grupo 1
X1 = np.random.rand(n//2, dimensiones) * 2
X1[:,0] += 1
X1[:,1] += 1

# Grupo 2
X2 = np.random.rand(n//2, dimensiones) * 2
X2[:,0] -= 1
X2[:,1] -= 1

# Juntamos los dos grupos
X = np.concatenate((X1, X2))


# imprimimos los datos generados con matplotlib
plt.figure()
plt.title("Datos generados")
plt.scatter(X[:,0], X[:,1])
plt.show()

# Guardamos los datos en un archivo
df = pd.DataFrame(X, columns=["x", "y"])
df.to_csv("data/datos.csv", index=False)
```

```python
# Generamos un conjunto de datos de ejemplo

# Número de datos
n = 1000
dimensiones = 2

# Generamos n puntos distribuidos gaussianamente en 2 grupos
# Grupo 1
X1 = np.random.randn(n//2, dimensiones) * 0.5
X1[:,0] += 1
X1[:,1] += 1

# Grupo 2
X2 = np.random.randn(n//2, dimensiones) * 0.5
X2[:,0] -= 1
X2[:,1] -= 1

# Juntamos los dos grupos
X = np.concatenate((X1, X2))

# imprimimos los datos generados con matplotlib
plt.figure()
plt.title("Datos generados")
plt.scatter(X[:,0], X[:,1])
plt.show()

# Guardamos los datos en un archivo
df = pd.DataFrame(X, columns=["x", "y"])
df.to_csv("data/datos_gaussianos.csv", index=False)
```

```python
#Generamos un conjunto de datos con 3 clusters distribuidos en forma de anillo
n = 1000
dimensiones = 3

# Generamos n puntos distribuidos uniformemente en 2 anillos, con ruido
# Grupo 1
theta = np.random.rand(n//2) * 2 * np.pi
X1 = np.zeros((n//2, dimensiones))
X1[:,0] = np.cos(theta)
X1[:,1] = np.sin(theta)
X1[:,0] += 0.5
X1[:,2] = np.random.rand(n//2) * 0.1

# Grupo 2
theta = np.random.rand(n//2) * 2 * np.pi
X2 = np.zeros((n//2, dimensiones))
X2[:,0] = np.cos(theta)
X2[:,1] = np.random.rand(n//2) * 0.1
X2[:,2] = np.sin(theta)
X2[:,0] -= 0.5

# Juntamos los dos grupos
X = np.concatenate((X1, X2))

# imprimimos los datos generados con matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.title("Datos generados")
ax.scatter(X[:,0], X[:,1], X[:,2])
plt.show()

# Guardamos los datos en un archivo
df = pd.DataFrame(X, columns=["x", "y", "z"])
df.to_csv("data/datos_anillo.csv", index=False)
```

```python
#generamos un conjunto de datos con 3 clusters distribuidos en forma de anillo concéntricos con ruido
n = 1000
dimensiones = 2

# Generamos n puntos distribuidos uniformemente en 3 anillos, con ruido
# Grupo 1
theta = np.random.rand(n//6) * 2 * np.pi
X1 = np.zeros((n//6, dimensiones))
X1[:,0] = np.cos(theta)*(0.05+ np.random.rand(n//6)*0.01)
X1[:,1] = np.sin(theta)*(0.05+ np.random.rand(n//6)*0.01)

# Grupo 2
theta = np.random.rand(n//3) * 2 * np.pi
X2 = np.zeros((n//3, dimensiones))
X2[:,0] = np.cos(theta)*(0.15+ np.random.rand(n//3)*0.01)
X2[:,1] = np.sin(theta)*(0.15+ np.random.rand(n//3)*0.01)

# Grupo 3
theta = np.random.rand((3*n)//6) * 2 * np.pi
X3 = np.zeros(((3*n)//6, dimensiones))
X3[:,0] = np.cos(theta)*(0.3+ np.random.rand((3*n)//6)*0.01)
X3[:,1] = np.sin(theta)*(0.3+ np.random.rand((3*n)//6)*0.01)


# Juntamos los tres grupos
X = np.concatenate((X1, X2, X3))

# imprimimos los datos generados con matplotlib
fig = plt.figure()
plt.title("Datos generados")
plt.scatter(X[:,0], X[:,1])
plt.show()

# Guardamos los datos en un archivo
df = pd.DataFrame(X, columns=["x", "y"])
df.to_csv("data/datos_anillo_concentrico.csv", index=False)
```

```python
#quiero generar 2 planos xy con ruido 
n = 500
dimensiones = 3

# Generamos n puntos distribuidos uniformemente en 2 planos xy, con ruido
# Grupo 1
X1 = np.random.rand(n//2, dimensiones)
X1[:,2] = -0.5+ np.random.rand(n//2) * 0.1 -0.05

# Grupo 2
X2 = np.random.rand(n//2, dimensiones) 
X2[:,2] = 0.5 + np.random.rand(n//2) * 0.1 - 0.05


# Juntamos los dos grupos
X = np.concatenate((X1, X2))

# imprimimos los datos generados con matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.title("Datos generados")
ax.scatter(X[:,0], X[:,1], X[:,2])
plt.show()

# Guardamos los datos en un archivo
df = pd.DataFrame(X, columns=["x", "y", "z"])
df.to_csv("data/datos_planos.csv", index=False)
```

```python
# vamos a generar varios grupos de datos en 2d

# un grupo de tamaño 500 con distribución normal centrado en (0.5,0.5) de radio 0.3
X1 = np.random.randn(500, 2) * 0.15
X1[:,0] += 1
X1[:,1] += 1


# un grupo de tamaño 200 con forma de rectangulo y distribución uniforme entre los puntos (0.0,0.0) y (0.4,0.5)
X2 = np.random.uniform(size=(200,2)) 
X2[:,0] += 0.0
X2[:,1] /= 3

# otro grupo de tamaño 300 con forma de rectangulo y distribución uniforme entre los puntos (0.6,0.5) y (1.0,1.0)
X3 = np.random.uniform(size=(300,2))
X3[:,0] /= 3
X3[:,1] += 0.5





#juntemos los dos grupos
X = np.concatenate((X1, X2, X3))

# imprimimos los datos generados con matplotlib
plt.figure()
plt.title("Datos generados")
plt.scatter(X[:,0], X[:,1])
plt.show()

# Guardamos los datos en un archivo
df = pd.DataFrame(X, columns=["x", "y"])
df.to_csv("data/datos_3grupos.csv", index=False)
```

```python
#vamos a usar la libreria sklearn para generar datos
from sklearn.datasets import make_blobs
#importaamos PCA para reducir la dimensionalidad
from sklearn.decomposition import PCA

muestras = 500
dimensiones = 2
clusters = 4
std = 0.3

#generamos 3 grupos de datos
X, y = make_blobs(n_samples=muestras, n_features=dimensiones, centers=clusters, cluster_std=std)

#reducimos la dimensionalidad a 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)


# imprimimos los datos generados con matplotlib
plt.figure()
plt.title("Datos generados")
plt.scatter(X_pca[:,0], X_pca[:,1])
plt.show()

# Guardamos los datos originales en un archivo
df = pd.DataFrame(X, columns=["x"+str(i) for i in range(dimensiones)])
df.to_csv("data/datos_sklearn"+str(dimensiones)+"d"+str(clusters)+"c.csv", index=False)
```

```python
# Número de puntos en cada clúster
n_points = 100

# Desviación estándar para los clústeres
std_dev = 0.1

# Coordenadas de los vértices de la pirámide
A = np.array([1, 0, 0])
B = np.array([-1, 0, 0])
C = np.array([0, np.sqrt(3), 0])
D = np.array([0, 0, 1])

# Generar datos alrededor de cada vértice
cluster_A = A + np.random.normal(0, std_dev, (n_points, 3))
cluster_B = B + np.random.normal(0, std_dev, (n_points, 3))
cluster_C = C + np.random.normal(0, std_dev, (n_points, 3))
cluster_D = D + np.random.normal(0, std_dev, (n_points, 3))

# Graficar los puntos
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(cluster_A[:,0], cluster_A[:,1], cluster_A[:,2], color='red', label='Clúster A')
ax.scatter(cluster_B[:,0], cluster_B[:,1], cluster_B[:,2], color='green', label='Clúster B')
ax.scatter(cluster_C[:,0], cluster_C[:,1], cluster_C[:,2], color='blue', label='Clúster C')
ax.scatter(cluster_D[:,0], cluster_D[:,1], cluster_D[:,2], color='purple', label='Clúster D')

# Configurar el gráfico
ax.set_title('Clústeres en los vértices de una pirámide triangular regular')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()
# Juntar todos los puntos
X = np.concatenate((cluster_A, cluster_B, cluster_C, cluster_D))

# Guardar los datos en un archivo
df = pd.DataFrame(X, columns=["x", "y", "z"])
df.to_csv("data/datos_piramide.csv", index=False)
```
